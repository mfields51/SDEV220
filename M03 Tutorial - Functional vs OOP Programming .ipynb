```python
# Sort an array of 0s, 1s, and 2s
def sort012(self, arr):
    low = 0
    middle = 0
    high = len(arr) - 1

    #Continue until all elements are sorted
    while middle <= high:
        if arr[middle] == 0:
            arr[low], arr[middle] = arr[middle], arr[low]
            low += 1
            middle += 1

        elif arr[middle] == 1:
            middle = middle + 1

        else:
            arr[middle], arr[high] = arr[high], arr[middle]
            high -= 1

    return arr
```

```python
#Binary Search
def binarysearch(self, arr, k):
    low = 0
    high = len(arr) - 1
    result = -1

    while low <= high:
        middle = (low + high) // 2

        if arr[middle] == k:
            result = middle
            high = middle - 1
        elif arr[middle] < k:
            low = middle + 1
        else:
            high = middle - 1

    return result
```
