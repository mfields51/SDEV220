```python
#assign '7' to the variable 'guess_me' and the value '1' to the variable 'number'
guess_me = 7
number = 1

#print 'too low' if low 
#print 'found it!' and exit once found
#print 'oops' and exit if too high
#incriment after end of each loop

while True:
    if number < guess_me:
        print("too low")
        number += 1
    elif number == guess_me:
        print("found it!")
        break
    else:
        print('oops')
        break
```
