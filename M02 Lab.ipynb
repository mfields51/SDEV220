## M02 Lab - Case Study: if...else and while
### The app will ask for last name to continue or 'ZZZ' to quit
### First name and GPA are then input
### Will use 'if' statement to decide if GPA qualifies for Dean's List (3.5+) or Honor Roll (3.25+)

```python
def main():

    while True:
        last_name = input("Enter student's last name or ZZZ to quit")
    
        if last_name.upper() == "ZZZ":
            break
    
        first_name = input("Enter student's first name")
        gpa = float(input("Enter student's GPA"))

        if gpa >= 3.5:
            print(f"{first_name} {last_name} has made the Dean's List!")
        elif gpa >= 3.25:
            print(f"{first_name} {last_name} has made Honor Roll!")
        else:
            print(f"{first_name} {last_name} does not qualify.")
    
if __name__ == "__main__":
    main()
```
