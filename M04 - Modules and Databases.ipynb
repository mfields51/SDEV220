```python
# 16.4 

# Use the sqlite3 module to create a SQLite database called books.db 
# and a table called books with these fields: title (text), author (text), and year (integer).
```

```python
# Use sqlite3
import sqlite3

# Create database called books.db
db = sqlite3.connect('books.db')
cursor = db.cursor()

# Add books (title:text, author:text, year:integer)
books_to_add = [
    ('The Hobbit', 'J.R.R. Tolkien', 1937),
    ('The Weirdstone of Brisingamen', 'Alan Garner', 1960),
    ('The Spellman Files', 'Lisa Lutz', 2007)
]

# Insert book list
cursor.executemany('INSERT INTO books VALUES (?, ?, ?)', books_to_add)

# Create table called books with these fields (title:text, author:text, and year:integer
cursor.execute('''CREATE TABLE IF NOT EXISTS books (
                    title TEXT,
                    author TEXT, 
                    year INTEGER)''')

# Commit and Close
db.commit()
db.close()
```

```python
# 16.8

# Use the sqlalchemy module to connect to the sqlite3 database books.db that you just made in exercise 16.4.
# select and print the title column from the book table in alphabetical order.
```

```python
# Use sqlalchemy
from sqlalchemy  import create_engine, text

# Connect books.db from 16.4
engine = create_engine('sqlite:///books.db')

# Select and print 
with engine.connect() as db:
    query = text("SELECT title FROM books ORDER BY title ASC")
    result = db.execute(query)

    # Print
    print("Titles in alaphabetical order: ")
    for row in result:
        print(row[0])
```
